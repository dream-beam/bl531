```python
import numpy as np
from ophyd import Device, Component, Signal

class VirtualCamera(Device):
    image = Component(Signal, value=np.zeros((240, 320)))
    width = Component(Signal, value=320)
    height = Component(Signal, value=240)
    
    def acquire(self):
        """Simulate acquiring an image."""
        self.image.put(np.random.random((self.height.get(), self.width.get())))
        return self.image.get()

# Create an instance
virtual_camera = VirtualCamera(name='virtual_camera')
```

```python
virtual_camera.acquire()
```
